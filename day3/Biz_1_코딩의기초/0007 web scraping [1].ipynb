웹을 표현하는 기본언어는 HTML이다. HTML을 분석하기 위한 대표적인 방법은 정규표현식(regular expression)이다. 정규표현식을 이용하여 HTML이외에도 다양한 유형의 텍스트에서 정보를 추출할 수 있다.

```
!pip install urllib # URL을 다루는 모듈 urllib을 설치
!pip install re # 문자를 표현하는 공식인 정규식표현식(Regular Expression) 모듈
```

```
import urllib.request # URL을 열기위한 함수와 클래스 등 정의 
import re

url ="http://finance.naver.com/item/main.nhn?code=051910" # 네이버 금융 LG화학 페이지 
html = urllib.request.urlopen(url) # URL을 연다 
html_contents =str(html.read().decode("ms949")) # MS의 한글 인코딩방식(ms949)

# 첫 번째 HTML 패턴
# [\s\S]+? --> 글자 전체를 의미
stock_results = re.findall("(\<dl class=\"blind\"\>)([\s\S]+?)(\<\/dl\>)", html_contents)
print(len(stock_results))
print('-------------------------------')

lg_stock_0 = stock_results[0]     # 두 개의 튜플 값 중 첫 번째 패턴
lg_stock_1 = stock_results[1]     # 두 개의 튜플 값 중 두 번째 패턴

lg_index_0 = lg_stock_0[1]     # 세 개의 튜플 값 중 두 번째 패턴
lg_index_1 = lg_stock_1[1]     # 세 개의 튜플 값 중 두 번째 패턴

# 주식 정보만 추출함
index_list_0= re.findall("(\<dd\>)([\s\S]+?)(\<\/dd\>)", lg_index_0)
index_list_1= re.findall("(\<dd\>)([\s\S]+?)(\<\/dd\>)", lg_index_1)

for index in index_list_0:
  # print(index[0])
  print(index[1])                # 세 개의 튜플 값 중 두 번째 값
  # print(index[2])

print('-------------------------------')
for index in index_list_1:
  # print(index[0])
  print(index[1])                # 세 개의 튜플 값 중 두 번째 값
  # print(index[2])
 
```
