```
import pandas as pd
import numpy as np
```

```
df1 = pd.DataFrame([10,20,20,20,30,30,40,50])
df1
```

```
## 간단한 코드로 데이터 프레임에 대한 평균을 구하여 본다.
df1.mean()  
```

```
## 평균에 대한 수학적 의미로 평균을 구하여 본다.

df1.sum()/len(df1)   
```

```
# 간단한 코드로 데이터 프레임에 대한 중앙값을 구한다
np.median(df1)
```

```
# 간단한 코드로 최빈값 구하기
df1.mode()
```

```
# 1 사분위 수
np.percentile(df1,25)
```

```
# 3 사분위 수
np.percentile(df1,75)
```

```
## 간단한 코드로 데이터 프레임에 대한 분산을 구하여 본다.

np.var(df1)  
```

```
## 분산에 대한 수학적 의미로 분산을 구하여 본다.

((df1 - df1.mean())**2).sum()/len(df1)  
```

```
## 간단한 코드로 데이터 프레임에 대한 표준편차를 구하여 본다.

np.std(df1)  
```

```
## 표준편차에 대한 수학적 의미로 표준편차를 구하여 본다

np.sqrt( ((df1 - df1.mean())**2).sum()/len(df1) ) 
```

```
import numpy as np
from keras.models import Sequential 
from keras.layers import Dense
```

```
X = np.array([[10, 20, 30], [20, 30, 40], [30, 40, 50], [40, 50, 60]])
y = np.array([40, 50, 60, 70])
```

```
model = Sequential()
model.add(Dense(100, activation='relu', input_dim=3, kernel_initializer='uniform'))
model.add(Dense(100, activation='relu', kernel_initializer='uniform'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
model.fit(X, y, epochs=1000, verbose=0)
x_input = np.array([50, 60, 70])
x_input = x_input.reshape((1, 3))
```

```
yhat = model.predict(x_input, verbose=0)
print(yhat[0,0])
```

```
# 데이터 변환(Data Transformation)
```

```
x = np.linspace(-5,5,100)
x
```

```
import matplotlib.pyplot as plt 

plt.plot(x)
```

```
##수학적 수식에 의한 데이터 변환 > Z_score Normalization(표준화)

z_score = (x-x.mean())/x.std()  
```

```
z_score
```

```
z_score.mean()
```

```
z_score.std()
```

```
plt.plot(z_score)
```

```
# 표준정규분포 식
def f(x):
    return 1/np.sqrt(2*np.pi)*np.exp(-x**2/2)

plt.plot(x,f(x))
```

```
import scipy.integrate as integrate

integrate.quad(f, -np.inf, 0 )[0]   # 표준정규분포에서 확률 구하기
```
