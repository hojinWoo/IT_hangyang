```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
#Generating Data
Ax=60+8*np.random.normal(size=100)
Ay=80+8*np.random.normal(size=100)
Bx=20+7*np.random.normal(size=88)
By=120+7*np.random.normal(size=88)
Cx=20+10*np.random.normal(size=66)
Cy=30+10*np.random.normal(size=66)

X=np.concatenate([Ax,Bx,Cx])
Y=np.concatenate([Ay,By,Cy])
#lable=> A:0,B:1,C:2
A=[0 for a in range(100)]
B=[1 for b in range(88)]
C=[2 for c in range(66)]
label=np.concatenate([A,B,C])

data=list(zip(X,Y)) # model.fit 받아주는 input type이 list라서 list로 변환
```

```python
fig, ax = plt.subplots()
ax.scatter(Ax,Ay, color='orange')
ax.scatter(Bx,By, color='turquoise')
ax.scatter(Cx,Cy, color='skyblue')
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,label,test_size=0.2,random_state=1)
```

```python
#instantiate the model
model=svm.SVC(kernel="linear",C=1) #min||w|| + C||error||

#fitting the model
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
print("accuracy : %f"%accuracy_score(pred,y_test))
```
