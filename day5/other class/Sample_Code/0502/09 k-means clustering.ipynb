```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Generating Data
Ax=60+8*np.random.normal(size=100)
Ay=80+8*np.random.normal(size=100)
Bx=20+7*np.random.normal(size=88)
By=120+7*np.random.normal(size=88)
Cx=20+10*np.random.normal(size=66)
Cy=30+10*np.random.normal(size=66)

X=np.concatenate([Ax,Bx,Cx])
Y=np.concatenate([Ay,By,Cy])
data=list(zip(X,Y))
```

```python
plt.scatter(X,Y)
```

```python
fig, ax = plt.subplots()
ax.scatter(Ax,Ay, color='orange')
ax.scatter(Bx,By, color='turquoise')
ax.scatter(Cx,Cy, color='skyblue')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
```

```python
#instantiate the model
kmeans=KMeans(n_clusters=3)
#fitting the model
kmeans.fit(data)

print("centers of clusters\n",kmeans.cluster_centers_)
print("labels of data\n",kmeans.labels_)
```

```python
#plot by labels
len_labels=len(kmeans.labels_)
Rx=[X[i] for i in range(len_labels) if kmeans.labels_[i]==0]
Ry=[Y[i] for i in range(len_labels) if kmeans.labels_[i]==0]
Gx=[X[i] for i in range(len_labels) if kmeans.labels_[i]==1]
Gy=[Y[i] for i in range(len_labels) if kmeans.labels_[i]==1]
Bx=[X[i] for i in range(len_labels) if kmeans.labels_[i]==2]
By=[Y[i] for i in range(len_labels) if kmeans.labels_[i]==2]
center_x=[i[0] for i in kmeans.cluster_centers_]
center_y=[i[1] for i in kmeans.cluster_centers_]

plt.figure(1)
plt.scatter(Rx,Ry,color='pink')
plt.scatter(Gx,Gy,color='lime')
plt.scatter(Bx,By,color='skyblue')
plt.scatter(center_x,center_y,marker="x",color="black",s=100)
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
```
