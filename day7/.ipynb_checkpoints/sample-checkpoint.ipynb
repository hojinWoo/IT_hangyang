```python
from pandas_datareader import data
import statsmodels.formula.api as sm
import pandas as pd
```

## import data

```python
SS = data.DataReader(name='005930.KS',data_source='yahoo',start='2000-01-01')
```

```python
SS.head()
```

```python
(0.082740-0.090599)/0.090599
```

```python
SS.loc[:,'Adj Close'].head()
```

```python
SS['ret']=SS['Adj Close'].pct_change(1)
```

```python
SS['Y']=(SS['ret'] >= 0)*1
```

```python
SS['X'] = SS['Volume'].pct_change(1)
```

```python
SS['lag_X']=SS['X'].shift(1)
```

```python
SS.head()
```

```python
import numpy as np
```

```python
SS2 = SS.replace([np.inf, -np.inf], np.nan).dropna(how='all')
```

```python
result = sm.ols(formula="Y ~ lag_X", data=SS2).fit()
print(result.summary())

#print(result.params)
#print(result.summary())
```
